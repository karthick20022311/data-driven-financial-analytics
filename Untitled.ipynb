```python
import os
import yaml
import pandas as pd
```

```python
DATA_FOLDER = r"C:\Users\karth\Downloads\data"
SECTOR_FILE = r"C:\Users\karth\Downloads\Sector_data - Sheet1.csv"
```

```python
sector_df = pd.read_csv(SECTOR_FILE)

# Check columns once
print("Columns:", sector_df.columns)

# Extract correct column
symbols = sector_df["Symbol"].tolist()

print("Total Symbols:", len(symbols))
print("Sample Symbols:", symbols[:5])
```

```python
all_records = []

for root, dirs, files in os.walk(DATA_FOLDER):
    for file in files:
        if file.endswith(".yaml"):
            
            file_path = os.path.join(root, file)
            date = file.replace(".yaml", "")
            
            with open(file_path, "r") as f:
                content = yaml.safe_load(f)

                for i, stock in enumerate(content):
                    record = {
                        "date": date,
                        "symbol": symbols[i],
                        "open": stock.get("open"),
                        "high": stock.get("high"),
                        "low": stock.get("low"),
                        "close": stock.get("close"),
                        "volume": stock.get("volume")
                    }
                    all_records.append(record)

df = pd.DataFrame(all_records)

print("Shape:", df.shape)
df.head()
```

```python
df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d_%H-%M-%S")

df.sort_values(["symbol", "date"], inplace=True)

df.to_csv("master_stock_data.csv", index=False)

print("Master dataset created successfully")
print("Final Shape:", df.shape)
print("Sample symbols:", df["symbol"].unique()[:5])
```

```python
# Extract only ticker after colon
df["symbol"] = df["symbol"].str.split(":").str[-1].str.strip()

print("Cleaned Symbols Sample:")
print(df["symbol"].unique()[:5])
```
